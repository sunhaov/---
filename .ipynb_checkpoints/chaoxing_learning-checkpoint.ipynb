# 使用 selenium
**超星学习通登录的验证码使用动态页面加载，加载页面之后会执行`init`函数，通过执行`getnumcode`获取数字验证码，而正常手段只能获取静态网页，不能执行网页内的函数，因此转而使用`selenium`操纵浏览器访问动态网页**

```python
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
 
chrome_options=Options()
#设置chrome浏览器无界面模式
# chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu') #谷歌文档提到需要加上这个属性来规避bug
browser = webdriver.Chrome(options=chrome_options)
browser.get('http://passport2.chaoxing.com/login?fid=&refer=http://i.mooc.chaoxing.com')
# print(browser.page_source)

#获取截图
browser.get_screenshot_as_file(r'C:\Users\Administrator\Desktop\screenshot.png')
browser.find_element_by_id("unameId").send_keys("13314138507")
browser.find_element_by_id("passwordId").send_keys("sunhao456123")



# 关闭浏览器
browser.close()  # 不是不打开，只是隐藏了，所以还是要关闭
# 关闭chreomedriver进程
browser.quit()
```

**居然 html 里嵌套 html，因此要再获取里边的 html**

```python
# 继续解析新的 html
with open('iframe.txt', 'r') as f:
    data = f.read()
print(data)
```

```python
soup = BeautifulSoup(data, features='lxml')
maogai = soup.find('div', {"class": "Mcon1img httpsClass"})
# print(maogai)
mg = str(maogai).split('"')[3]  # 转换为字符串并进行切分,取出想要的链接
# print(mg)
maog = mg.replace('&amp;', '&')  # 对比真正网站的 url 发现相比提取出来的 url，没有 amp;
print(maog)
```

# 把 selenium 获取到的 cookies 传给 requests

```python
import requests

# 获取 cookies 添加到 requests 里
cookie = browser.get_cookies()
c = requests.cookies.RequestsCookieJar()
for i in cookie:  # 添加cookie到CookieJar
    c.set(i["name"], i["value"])
# 新建 session
s = requests.session()
s.cookies.update(c)  # 更新session里的cookie

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.181 Safari/537.36'}
r = s.get(url=maog, headers=headers)
soup = BeautifulSoup(r.text, features='lxml')
# 这里 r.text 的内容跟上边 course.txt 的内容是一样的,这里只是保存记录一下
```

# 在毛概课页面内解析找到讨论组
**继续加油！**

```python
# 继续解析新的 html
with open('course.txt', 'r') as f:
    data = f.read()
print(data)
```

```python
soup = BeautifulSoup(data, features='lxml')
navshow = soup.find('div', {"class": "navshow"})
# print(navshow)
ul = navshow.find('ul')
for li in ul:
    if('讨论' in str(li)):
        discuss_li = li
        break
# print(discuss_li)
discuss = str(discuss_li).split('"')[1]
# print(discuss)
disc = discuss.replace('&amp;', '&')
dis = maog[:28] + disc  # 讨论组的网页
print(dis)
```

# 进入单个小组

```python
# 使用 listbox 获取选中的小组索引
select_index = (2, 3, 4, 5, 6, 7, 8, 10, 11, 12, 13, 14, 15)
# 先只看一个小组，九章11组，索引为3
# 这里就保存到本地 group.txt
with open('group.txt', 'r') as f:
    data = f.read()
# print(data)
```

```python
soup = BeautifulSoup(data, features='lxml')
reply_title = soup.find_all('h3', {'class': 'bt ol'})  # 所有标题
dic_grp = {}
# grp_topic = []
for i,t in enumerate(reply_title):
    if(i == 0):
        grp_title = t.get_text().strip()
        dic_grp[grp_title] = 0  # 0是随便写的,给问题留的地方
    else:
        break  # 不用存问题，直接 break 就行
#         grp_topic.append({t.get_text():0})  # 这个0是给成员回复数留的地方
# dic_grp[grp_title] = grp_topic  # 把 问题 存到 小组 对应的值
# print(dic_grp)
```

```python
name_span = soup.find_all('span', {'class': 'name'})
# 所有人和建贴人名字后边都有空格，盖楼人没有空格
names = [ns.get_text() for ns in name_span]
nozz_names = set(names)
nozz_names.remove(names[1])  # 然后在集合中去掉盖楼人,也就是names的第二个，没有空格的名字
fayan_num = [n + ': ' + str(names[1:].count(n)) for n in nozz_names]  # 第一个是建贴人，不要建贴人
dic_grp[grp_title] = fayan_num
print(dic_grp)
```
